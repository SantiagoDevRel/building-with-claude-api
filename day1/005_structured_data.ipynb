````python
## Response streaming
from dotenv import load_dotenv
from anthropic import Anthropic

load_dotenv()
client = Anthropic()

## Setup model
model = "claude-sonnet-4-0"
messages_list = []

## Add user & assistant msgs
def add_user_message(messages_list, msg):
    formatted_user_msg = {"role":"user", "content":msg}
    messages_list.append(formatted_user_msg)

def add_assistant_message(messages_list, msg):
    formatted_assistant_msg = {"role":"assistant", "content":msg}
    messages_list.append(formatted_assistant_msg)  

# Chat
def chat(messages_list, system = None, temperature = 1.0, stop_sequences = None):
    params = {
        "model" : model,
        "max_tokens":  1000,
        "messages" : messages_list,
        "temperature":temperature,
    }

    if system:
        params["system"] = system

    if stop_sequences:
        params["stop_sequences"] = stop_sequences

    message = client.messages.create(**params)
    return message.content[0].text

add_user_message(messages_list, "create a json request for ethereum request balance in json")

""" prefilling the response from claude so it starts from ```json """
add_assistant_message(messages_list, "```")

response = chat(messages_list, stop_sequences=["```"])
response.strip()
````
