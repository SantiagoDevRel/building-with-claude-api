```python
from dotenv import load_dotenv
from anthropic import Anthropic

load_dotenv()
client = Anthropic()

## Setup model
model = "claude-sonnet-4-0"
messages_list = []

## Define functions
## Chat
def chat(messages_list, system = None, temperature = 1.0):
    params = {
        "model" : model,
        "max_tokens":  1000,
        "messages" : messages_list,
        "temperature":temperature
    }

    if system:
        params["system"] = system

    message = client.messages.create(**params)
    return message.content[0].text

## Add user & assistant msgs
def add_user_message(messages_list, msg):
    formatted_user_msg = {"role":"user", "content":msg}
    messages_list.append(formatted_user_msg)

def add_assistant_message(messages_list, msg):
    formatted_assistant_msg = {"role":"assistant", "content":msg}
    messages_list.append(formatted_assistant_msg)  

## Input function
while True:
   user_input = input("You: ")
   # Type quit to exit the loop
   if user_input.lower() == "quit":
        break

   add_user_message(messages_list, user_input)
   system = "" ## ADD SYSTEM!!
   response = chat(messages_list, system)
   add_assistant_message(messages_list, response)
   print("-----")
   print(f"response {(len(messages_list) / 2)} \n{response}")
   print("-----")
```
