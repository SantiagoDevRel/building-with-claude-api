```python
## Load env variables

from dotenv import load_dotenv

load_dotenv()
```

```python
## Create API client

from anthropic import Anthropic

client = Anthropic()

model = "claude-sonnet-4-0"
```

```python
def add_user_message(messages, text):
    user_message = {"role":"user","content":text}
    messages.append(user_message)

def add_assistant_message(messages, text):
    assistant_message = {"role":"assistant","content":text}
    messages.append(assistant_message)

def chat(messages):
    message = client.messages.create(
        model = model,
        max_tokens=1000,
        messages= messages
    )
    return message.content[0].text
```

```python
import json

## Make a starting list of messages
messages = []

## Add in the initial user questions
question_1 = "whats the best thing about becoming an AI DevRel? reply in 1 sentence"
add_user_message(messages, question_1)
print(f"Added first msg-> \n{messages}")

## Pass the list of msgs into chat to get an answer
answer_1 = chat(messages)
print(f"show answer1-> \n{answer_1}")

## Take the answer and add it to the add_assistant_message()
add_assistant_message(messages, answer_1)
print(f"add first answer to msgs-> \n{messages}")

## Add the user follow up question
question_2 = "tell me another sentence about it"
add_user_message(messages, question_2)
print(f"added question 2-> \n{messages}")

answer_2 = chat(messages)
print(f"answer 2-> {answer_2}")

add_assistant_message(messages, answer_2)
print(json.dumps(messages, indent=2))

## Print final answer
print(f"Final answer from messages-> {messages[-1]["content"]}")
```
